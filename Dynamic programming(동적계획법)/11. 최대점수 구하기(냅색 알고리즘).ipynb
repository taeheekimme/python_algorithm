이번 정보올림피아드대회에서 좋은 성적을 내기 위하여 현수는 선생님이 주신 N개의 문제를 풀려고 합니다. <br>
각 문제는 그것을 풀었을 때 얻는 점수와 푸는데 걸리는 시간이 주어지게 됩니다. <br>
제한시간 M안에 N개의 문제 중 최대점수를 얻을 수 있도록 해야 합니다. <br>
(해당문제는 해당시간이 걸리면 푸는 걸로 간주한다, 한 유형당 한개만 풀 수 있습니다.)


▣ 입력설명<br>
첫 번째 줄에 문제의 개수N(1<=N<=100)과 제한 시간 M(10<=M<=1000)이 주어집니다.<br>
두 번째 줄부터 N줄에 걸쳐 문제를 풀었을 때의 점수와 푸는데 걸리는 시간이 주어집니다.


▣ 출력설명<br>
첫 번째 줄에 제한 시간안에 얻을 수 있는 최대 점수를 출력합니다.


▣ 입력예제 1<br>
5 20<br>
10 5<br>
25 12<br>
15 8<br>
6 3<br>
7 4<br>
- 5가지 문제 유형, 제한시간 20
- 10점짜리 문제 1개를 5분안에 풀어야 함<br><br>
- 5가지 문제 유형, 제한시간 20으로 인덱스 0을 포함한 6행 21열 **2차원 리스트** 생성
- dy[i][j] : i번째 문제까지 적용했을 때,j분 안에 풀어 얻을 수 있는 최대 점수
- dy[3][15] : 3번째 문제 15 8까지 적용했을 때, 15분 안에 풀어 얻을 수 있는 최대 점수 입력
- 위와 같이 기존 방식으로 풀게 되면 한 유형 당 여러문제를 풀게 되어 **문제 조건에 위배**<br><br>

- dy[i-1][j-pt] : 현재 적용된 것을 적용하면 중복되므로 -1, 현재 적용되고 있는 문제 유형 시간 pt를 빼줌
- 행마다 나온 결과값을 다음 행에 복사하여 업데이트

- 실전에서는 2차원 리스트를 풀었을 때, 시간이 오래거리므로 **1차원 리스트**로 풀 것
- dy[j] : j분이 주어졌을 때, 얻을 수 있는 최대 점수
- 10 5 : pv pt
- 앞에서부터 순서대로 하면 문제 유형이 중복되므로 **뒤**에서부터 진행
    - dy[20] : dy[20-5] +10 =10
    - dy[19] : dy[19-5] +10 =10
    - dy[18] : dy[18-5] +10 =10
    - dy[17] : dy[17-5] +10 =10
    - dy[16] : dy[16-5] +10 =10
    - dy[15] : dy[15-5] +10 =10
    - dy[14] : dy[14-5] +10 =10
    - dy[13] : dy[13-5] +10 =10
    - dy[12] : dy[12-5] +10 =10
    - dy[11] : dy[11-5] +10 =10
    - dy[10] : dy[10-5] +10 =10
    - dy[9] : dy[9-5] +10 =10
    - dy[8] : dy[8-5] +10 =10
    - dy[7] : dy[7-5] +10 =10
    - dy[6] : dy[6-5] +10 =10
    - dy[5] : dy[5-5] +10 =10<br><br>
- 25 12 : pv pt
    - dy[20] : dy[20-12] +25 = dy[8]의 값 10 + 25 = 35는 dy[20]값인 10보다 높으므로 갱신
    - dy[19] : dy[19-12] +25 = dy[7]의 값 10 + 25 = 35는 dy[19]값인 10보다 높으므로 갱신
    - dy[18] : dy[18-12] +25 = dy[6]의 값 10 + 25 = 35는 dy[18]값인 10보다 높으므로 갱신
    - dy[17] : dy[17-12] +25 = dy[5]의 값 10 + 25 = 35는 dy[17]값인 10보다 높으므로 갱신
    - dy[16] : dy[16-12] +25 = dy[4]의 값 0 + 25 = 25는 dy[16]값인 10보다 높으므로 갱신
    - dy[15] : dy[15-12] +25 = dy[3]의 값 0 + 25 = 25는 dy[15]값인 10보다 높으므로 갱신
    - dy[14] : dy[14-12] +25 = dy[2]의 값 0 + 25 = 25는 dy[14]값인 10보다 높으므로 갱신
    - dy[13] : dy[13-12] +25 = dy[1]의 값 0 + 25 = 25는 dy[13]값인 10보다 높으므로 갱신
    - dy[12] : dy[12-12] +25 = dy[0]의 값 0 + 25 = 25는 dy[12]값인 10보다 높으므로 갱신<br><br>
    
- 나머지도 계산 한 후, 큰 값 출력

▣ 출력예제 1<br>
41

```python
if __name__=="__main__":
    n, m=map(int, input().split()) # 문제 유형 5가지, 제한시간 20
    dy=[0]*(m+1); # 0으로 초기화해 인덱스 m까지 나오도록 m+1
    
    for i in range(n): # 문제 유형 입력
        ps, pt=map(int, input().split()) # 문제 풀어서 얻을 수 있는 점수, 문제 푸는데 걸리는 시간
        
        for j in range(m, pt-1, -1): # m부터 p까지 뒤에서부터 시작하도록 -1
            dy[j]=max(dy[j], dy[j-pt]+ps) # 기존 j값과 j에서 현재 주어진 시간 pt를 뺀 값 중 최대값 선택
            
    print(dy[m])
```
